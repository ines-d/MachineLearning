## 5.1 regression linéaire univariée avec Numpy

```python
x=car_data.Year
y=car_data.Selling_Price
plt.scatter(x, y)
```

```python
import numpy as np

# pour avoir a et b de (ax +b) donc les coeff de theta
fit = np.polyfit(x,y, 1)
# pour avoir la fonction correspondante
poly = np.poly1d(fit)
#afficher les graphiques
plt.scatter(x, y)
plt.plot(x, poly(x), c='r')
```

## 5.2 regression linéaire univariée avec Scipy

```python
import scipy 
from scipy import stats

# renvoie un tuple avec 5 valeurs:la pente,l'ordonnée à l'origine, le coefficient de corrélation positif ou négatif,
#la p-value, l'erreur standard de l'estimation du gradient.
A = scipy.stats.linregress(x,y)
A
```

```python
# créer une fonction pour calculer la prediction 
def Predict(x):
   return A.slope * x + A.intercept
```

```python
# on créer une variable nomée fitLine (ligne d'ajustement) qui correspond a n otre fonction predict
fitLine = Predict(x)
# afficher les grapfiques 
plt.scatter(x, y)
plt.plot(x, fitLine, c='r')
```
